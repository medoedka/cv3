```python
rot_matrix = cv2.getRotationMatrix2D(
    center=(img.shape[1] / 2, img.shape[0] / 2),
    angle=90,
    scale=1
)
result = cv2.warpAffine(
    src=img,
    M=rot_matrix,
    dsize=img.shape[1::-1],
    flags=cv2.INTER_NEAREST,
    borderMode=cv2.BORDER_CONSTANT,
    borderValue=123.5
)
plt.imshow(result)
```
